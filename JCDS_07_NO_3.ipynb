```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
```

```python
web = requests.get("https://www2.2019seagames.com/countries/")
data = BeautifulSoup(web.content, "html.parser")

d2017 = []

for i in data.find_all("em"):
    d2017.append(i.text)

d2017
```

```python
negara = d2017[::6]

negara = negara[0:11]

negara
```

```python
gold = d2017[2::6]

gold
```

```python
gold1 = []
goldfiks = []

for i in range(len(gold)):
    hasil = gold[i].split(" ")
    gold1.append(hasil[1])
    
for j in range(len(gold1)):
    hasil1 = int(gold1[j])
    goldfiks.append(hasil1)
    
goldfiks = goldfiks[:11]

goldfiks
```

```python
web1 = requests.get("https://rs.2019seagames.com/RS2019/mobiapp/MedalTally")
data1 = BeautifulSoup(web1.content, "html.parser")

out1 = []

for i in data1.find_all("small"):
    out1.append(i.text)

out1
out2 = out1[5:]
```

```python
negara19 = out2[::5]

negara19
```

```python
gold1 = out2[1::5]

gold1
```

```python
all0 = list(zip(negara, goldfiks))
all0.sort()

all1 = list(zip(negara19, gold1))
all1.sort()
```

```python
all0
```

```python
all1
```

```python
country_all = []

for i in range(len(all1)):
    temp = all1[i]
    country_all.append(temp[0])

country_all
```

```python
gold_2017 = []

for i in range(len(all0)):
    temp = all0[i]
    gold_2017.append(temp[1])

gold_2017
```

```python
gold_2019 = []

for i in range(len(all1)):
    temp = all1[i]
    gold_2019.append(int(temp[1]))

gold_2019
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,7))
plt.plot(country_all, gold_2017, "r-o")
plt.plot(country_all, gold_2019, "g-o")
plt.plot(country_all[gold_2017.index(max(gold_2017))], max(gold_2017), '*', markersize = 20, markerfacecolor = "navy")
plt.plot(country_all[gold_2019.index(max(gold_2019))], max(gold_2019), '*', markersize = 20, markerfacecolor = "navy")
plt.title('GOLD MEDAL TALLY SEA GAMES 2017 & 2019', fontsize = 20, pad = 60)
plt.xticks(rotation = 25)
plt.legend(['Gold 2017', 'Gold 2019', 'Champion 2017', 'Champion 2019'])
plt.show
```

```python
plt.figure(figsize = (15,7))
plt.subplot(1, 2, 1)
plt.pie(gold_2017, labels = country_all, colors=['#381460','#b2dffb','#e7a4e4','#ffc55c','#f6c3e5','#a278b5','#381460','#d77fa1','#e6b2c6','#ad62aa','#e78fb3'],autopct='%1.1f%%')
plt.title('Gold% Medal Tally SEA GAMES 2017', fontsize = 20)

plt.subplot(1, 2, 2)
plt.pie(gold_2019, labels = country_all,colors=['#381460','#b2dffb','#e7a4e4','#ffc55c','#f6c3e5','#a278b5','#381460','#d77fa1','#e6b2c6','#ad62aa','#e78fb3'], autopct='%1.1f%%')
plt.title('Gold% Medal Tally SEA GAMES 2019', fontsize = 20)
plt.show()
```
